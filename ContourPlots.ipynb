Step 1: import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

import matplotlib.cm as cm
import matplotlib.mlab as mlab
```

A few function definitions

A routine to drwa a contour plot for a field phi (see https://matplotlib.org/examples/pylab_examples/contour_demo.html)

```python
def cont(phi):
    delta = 0.025
    x = np.arange(-10.0, 10.0, delta)
    y = np.arange(-10, 10.0, delta)
    X, Y = np.meshgrid(x, y)

    Z = phi(X,Y)

    # Create a simple contour plot with labels using default colors.  The
    # inline argument to clabel will control whether the labels are draw
    # over the line segments of the contour, removing the lines beneath
    # the label
    CS = plt.contour(X, Y, Z)
```

```python
def showCont(phi):
    plt.figure(figsize=(5,5))
    cont(phi)
    plt.show
```

Below, draw a contour plot for $\phi = -e^{-\frac{x^2 + y^2}{2\sigma^2}}$

```python
def phiquizzB(X, Y):
    epsilon = 0.000001
    sigma=5
    return -np.exp(-(X**2 + Y**2)/(2*sigma**2))
showCont(phiquizzB)
```

Below: $\phi = x - y $;

```python
def phiquizzC(X, Y):
    return X - Y
showCont(phiquizzC)
```

Below: $\phi = x + y^2$;

```python
def phi2(X, Y):
    return X + Y**2
showCont(phi2)
```

Below: $\phi = \frac{1}{r} = \frac{1}{\sqrt{x^2 + y^2}}$. The contours for $\phi$ are only drawn for $r > 1$. 

(Change minRadius in the code below to see what happens.)

```python
def phi3(X, Y):
    minRadius = 1;
    # phi goes to infinity at 0,0, and goes very quickly very large near there; 
    # while the usual "add epsilon" trick avoids the code crash at (0,0), the
    # problem remains that phi goes very large very quickly near (0,0) and 
    # effectively all the contours shrink to a little circle 
    # (remember we draw contours at equal height intervals). Therefore, we
    # need to exclude central region to get decent plot.
    r = np.maximum(minRadius, np.sqrt(X**2 + Y**2))
    return 1/r 
showCont(phi3)
```

Below: $\phi = y$;

```python
def phi5(X, Y):
    return Y
showCont(phi5)
```

Below:$\phi = -x$

```python
def phi7(X, Y):
    return -X
showCont(phi7)
```

Below: $\phi = -xy$;

```python
def phi9(X, Y):
    return -X*Y
showCont(phi9)
```
