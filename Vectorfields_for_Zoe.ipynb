Step 1: import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

import matplotlib.cm as cm
import matplotlib.mlab as mlab
```

let's take as our first field: phi = $-(x^2/a + y^2/b)$ with a=2, b=4, gives F=(x, y/2)

```python
def vectors(F):
    X, Y = np.meshgrid(np.arange(-10, 11, 2), np.arange(-10, 11, 2))
    U, V = F(X, Y) ##X
        #V = Fy(X, Y) ##Y/2

    #plt.figure(figsize=(5,5))
#plt.title('Arrows scale with plot width, not view')
    Q = plt.quiver(X, Y, U, V, units='width')
    #qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E', coordinates='figure')

#    plt.scatter(X, Y, color='k', s=2)
```

```python
def showVec(F):
    plt.figure(figsize=(5,5))
    vectors(F)
    plt.show()
```

# Fields with non-zero divergence that do not spread spread out

```python
def F2a(X, Y):
    return X*X+100, 0
showVec(F2a)
#divergence = 2x
```

```python
def F2b(X, Y):
    return X-Y, Y-X
showVec(F2b)
#divergence = 2
```

```python
def F2c(X, Y):
    return np.cos(X)**2, 0
showVec(F2c)
#divergence = -sin(X)
```

```python
def F2d(X, Y):
    return 0, Y+40
showVec(F2d)
#divergence = 1
```

# Fields with zero divergence that I think looks like they spread out:

```python
def F1b(X, Y):
    return (X-12)*(X-12), -2*(X-12)*Y
showVec(F1b)
#divergence = 2(X-12)-2(X-12)=0
```

```python
def F1a(X, Y):
    return X+Y, X-Y
showVec(F1a)
#divergence = 1-1=0
```
