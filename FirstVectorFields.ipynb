# vector fields and contour plots

Step 1: import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

import matplotlib.cm as cm
import matplotlib.mlab as mlab
```

A function that plots a vector field F:

```python
def vectors(F):
    X, Y = np.meshgrid(np.arange(-10, 11, 2), np.arange(-10, 11, 2))
    U, V = F(X, Y) ##X
        #V = Fy(X, Y) ##Y/2

    plt.figure(figsize=(5,5))
#plt.title('Arrows scale with plot width, not view')
    Q = plt.quiver(X, Y, U, V, units='width')
    #qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E', coordinates='figure')

    #plt.scatter(X, Y, color='k', s=2)

    plt.show()
```

## Let's draw a few vector fields

(I'll be writing the mathematical expression for the same vector field in different ways to get you used to the different ways of expressing the same thing that we will use).

Below: $\vec{F}_1(x,y) = x\, \vec{\imath} + y \,\vec{\jmath} = \left(\begin{array}{c} 
    x \\ 
    y 
    \end{array}\right) = \sqrt{x^2 + y^2} \, \vec{e}_r = r\,\vec{e}_r $

```python
def F1(X, Y):
    return X, Y
vectors(F1)
```

Below: $\vec{F}_2(x,y) = y\, \vec{\imath} + x \,\vec{\jmath} = \left(\begin{array}{c} 
    y \\ 
    x 
    \end{array}\right)$

```python
def F2(X, Y):
    return Y, X
vectors(F2)
```

Below: $\vec{F}_3(x,y) = y\, \vec{\imath} + x \,\vec{\jmath} = \left(\begin{array}{c} 
    -y \\ 
    x 
    \end{array}\right) = \sqrt{x^2 + y^2}\, \vec{e}_{\phi} = r \,\vec{e}_{\phi}$

```python
def F3(X, Y):
    return -Y, X
vectors(F3)
```

Below: $\vec{F}_4(x,y) = \frac{1}{\sqrt{x^2 + y^2}} \left(y\, \vec{\imath} + x \,\vec{\jmath}\right) 
= \frac{y}{\sqrt{x^2 + y^2}}\, \vec{\imath} + \frac{x}{\sqrt{x^2 + y^2}} \,\vec{\jmath}
=
\frac{1}{\sqrt{x^2 + y^2}}
\left(\begin{array}{c} 
    -y \\ 
    x 
    \end{array}\right) =  \vec{e}_{\phi}$
  
(In this field, all vectors have the same length; ignore the dot in the centre)

```python
def F4(X, Y):
    epsilon = 0.0000001
    # (adding epsilon to denominator to avoid division by zero)
    return -Y/np.sqrt(X**2 + Y**2 + epsilon), X/np.sqrt(X**2 + Y**2 + epsilon)
vectors(F4)
```

Below: $\vec{F}_5(x,y) = 2x\, \vec{\imath} + y \,\vec{\jmath} = \left(\begin{array}{c} 
    2x \\ 
    y
    \end{array}\right)$

```python
def F5(X, Y):
    return 2*X, Y
vectors(F5)
```

Below: $\vec{F}_6(x,y) = x\, \vec{\imath} + 2y \,\vec{\jmath} = \left(\begin{array}{c} 
    x \\ 
    2y
    \end{array}\right)$

```python
def F6(X, Y):
    return X, 2*Y
vectors(F6)
```

Below: $\vec{F}_7(x,y) = \frac{x}{(x^2 + y^2)}\, \vec{\imath} + \frac{y}{(x^2 + y^2)} \,\vec{\jmath} = 
    \frac{1}{x^2 + y^2}
    \left(\begin{array}{c} 
    x \\ 
    y
    \end{array}\right) = \frac{1}{r}\, \vec{e}_r $

```python
def F7(X, Y):
    epsilon=0.000000001
    return X/(X**2 + Y**2 + epsilon), Y/(X**2 + Y**2 + epsilon)
vectors(F7)
```

Below: $\vec{F}_8(x,y) = \frac{x}{(x^2 + y^2)^{3/4}}\, \vec{\imath} + \frac{y}{(x^2 + y^2)^{3/4}} \,\vec{\jmath} = 
    \frac{1}{(x^2 + y^2)^{3/4}}
    \left(\begin{array}{c} 
    x \\ 
    y
    \end{array}\right) = \frac{1}{\sqrt{r}}\, \vec{e}_r $

```python
def F8(X, Y):
    epsilon=0.000000001
    return X/(X**2 + Y**2 + epsilon)**(3/4), Y/(X**2 + Y**2 + epsilon)**(3/4)
vectors(F8)
```
