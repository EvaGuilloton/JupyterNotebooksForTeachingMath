# Plotting paths in 3D

```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

import matplotlib.cm as cm
import matplotlib.mlab as mlab
import mpl_toolkits.mplot3d as plt3
from mpl_toolkits.mplot3d import axes3d

from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

Following: https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

```python
def screwpath(t):
    r = 1
    x = r * np.sin(t)
    y = r * np.cos(t)
    z = t
    name = '(r*sin(t), r*cos(t), t)'
    return x, y, z, name
        
def multiplot3D(paths, mima, steps=400):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    t = np.linspace(mima[0], mima[1], steps)
    for path in paths:
        x, y, z, name = path(t)
        ax.plot(x, y, z, label=name)
    ax.legend()
    ax.set_xlabel('$X$')
    ax.set_ylabel('$Y$')
    ax.set_zlabel('$Z$')
    plt.show()
    
def plotpath(path, mima, steps=400):
    multiplot3D((path,), mima, steps)
    
plotpath(screwpath, (-4*np.pi, 4*np.pi))
```

```python
def straightLine(t):
    a=1
    b=2
    c=5
    x = a*t
    y = b*t
    z = c*t
    name = "straight line"
    return x, y, z, name
plotpath(straightLine, (0, 1))
```

```python
def MatchingAssignment_1(t):
    a=1
    b=2
    x = a * np.cos(t)
    y = t*0+2  # weird way of writing y=2
    z = b * np.sin(t)
    name = '(a*cos(t), 2, b*sin(t))'
    return x, y, z, name

plotpath(MatchingAssignment_1, (-np.pi, np.pi))
```

```python
def MatchingAssignment_2(t):
    a=1
    b=2
    x = t
    y = t*0+2  # weird way of writing y=2
    z = b * (1 - t**2/a**2)
    name = '(t, 2, b(1 - t^2/a^2))'
    return x, y, z, name

plotpath(MatchingAssignment_2, (-10, 10))
```

```python
def MatchingAssignment_3(t):
    a=1
    b=2
    x = -a*b/t
    y = t
    z = t**2
    name = '(-ab/t, t, t^2)'
    return x, y, z, name

plotpath(MatchingAssignment_3, (1, 10))
```

```python
def MatchingAssignment_4(t):
    r = 1
    x = r * np.sin(t)
    y = t
    z = r * np.cos(t)
    name = '(r*sin(t), t, r*cos(t))'
    return x, y, z, name
plotpath(MatchingAssignment_4, (-4*np.pi, 4*np.pi))
```

```python
def MatchingAssignment_5a(t):
    r = 1
    x = r * np.sinh(t)
    y = t
    z = r * np.cosh(t)
    name = '(r*sinh(t), t, r*cosh(t))'
    return x, y, z, name
def MatchingAssignment_5b(t):
    r = 1
    x = r * np.sinh(t)
    y = t
    z = -r * np.cosh(t)
    name = '(r*sinh(t), t, -r*cosh(t))'
    return x, y, z, name

multiplot3D( (MatchingAssignment_5a,MatchingAssignment_5b), (-4, 4))
```
