# vector fields and contour plots

Step 1: import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

import matplotlib.cm as cm
import matplotlib.mlab as mlab
```

let's take as our first field: phi = $-(x^2/a + y^2/b)$ with a=2, b=4, gives F=(x, y/2)

```python
def vectors(F):
    X, Y = np.meshgrid(np.arange(-10, 11, 2), np.arange(-10, 11, 2))
    U, V = F(X, Y) ##X
        #V = Fy(X, Y) ##Y/2

    #plt.figure(figsize=(5,5))
#plt.title('Arrows scale with plot width, not view')
    Q = plt.quiver(X, Y, U, V, units='width')
    #qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E', coordinates='figure')

#    plt.scatter(X, Y, color='k', s=2)
```

Now the correspoinding contour plot (see https://matplotlib.org/examples/pylab_examples/contour_demo.html)

```python
def cont(phi):
    delta = 0.025
    x = np.arange(-10.0, 10.0, delta)
    y = np.arange(-10, 10.0, delta)
    X, Y = np.meshgrid(x, y)

    Z = phi(X,Y)

    # Create a simple contour plot with labels using default colors.  The
    # inline argument to clabel will control whether the labels are draw
    # over the line segments of the contour, removing the lines beneath
    # the label
    CS = plt.contour(X, Y, Z)
```

```python
def showVec(F):
    plt.figure(figsize=(5,5))
    vectors(F)
    plt.show()
```

```python
def showCont(phi):
    plt.figure(figsize=(5,5))
    cont(phi)
```

```python
def showPlots(F, phi):
    # shows plots seperately, and combined.
    showVec(F)
    showCont(phi)
    plt.figure(figsize=(5,5))
    vectors(F)
    cont(phi)
    plt.show()
```

```python
def FquizzB(X, Y):
    epsilon = 0.000001
    sigma=5
    c=np.exp(-(X**2 + Y**2)/(2*sigma**2))
    return X/sigma**2 * c, Y/sigma**2 * c
def phiquizzB(X, Y):
    epsilon = 0.000001
    sigma=5
    return -np.exp(-(X**2 + Y**2)/(2*sigma**2))
showPlots(FquizzB, phiquizzB)
```

```python
def FquizzC(X, Y):
    return X/2, Y
def phiquizzC(X, Y):
    return -X**2/4 - Y**2/2
showPlots(FquizzC, phiquizzC)
```

```python
def F2(X, Y):
    return 1, - np.sin(Y)
def phi2(X, Y):
    return X + np.cos(Y)
showPlots(F2, phi2)
```

```python
def F2(X, Y):
    eps = 0.000001
    return -Y/np.sqrt(X**2+Y**2 + eps), X/np.sqrt(X**2+Y**2 + eps)
showVec(F2)
```

```python
def FquizzA(X, Y):
    epsilon=0.0000001
    return -Y, X/(0.2*np.sqrt(X**2 + Y**2 + epsilon))
showVec(FquizzA)
```

```python
def F3(X, Y):
    eps = 0.000001
    return X/np.sqrt(X**2 + Y**2 + eps)**3, Y/np.sqrt(X**2 + Y**2 + eps)**3
def phi3(X, Y):
    eps = 1
    return 1/np.sqrt(X**2 + Y**2 + eps)
showPlots(F3, phi3)
```
