# Plotting paths in 2D

```python
##%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
```

```python
def circle(t):
    r=1
    x = r * np.sin(t)
    y = r * np.cos(t)
    name = 'a circle: (r*sin(t), r*cos(t))'
    return x, y, name

    
def multiplot2D(paths, mima, steps=400, title=""):
    fig = plt.figure()
    t = np.linspace(mima[0], mima[1], steps)
    name='lastname'
    for path in paths:
        x, y, name = path(t)
        plt.axis('equal')
        plt.plot(x, y, label=name)
        
    if(len(paths) == 1 and title == ""):
        plt.title(name)
    else:
        plt.legend(loc='upper left')
        plt.title(title)
    plt.show()
    
def plotpath2D(path, mima, steps=400):
    multiplot2D((path,), mima, steps)
    
    
plotpath2D(circle, (-4*np.pi, 4*np.pi))
```

```python
def straightLine(t):
    a=1
    b=2
    x = a*t
    y = b*t
    name = "straight line"
    return x, y, name
plotpath2D(straightLine, (0, 1))
```

```python
def MatchingAssignment_1(t):
    a=1
    b=2
    x = a * np.cos(t)
    z = b * np.sin(t)
    name = 'An ellipse: (a*cos(t), b*sin(t))'
    return x, z, name

plotpath2D(MatchingAssignment_1, (-np.pi, np.pi))
```

```python
def MatchingAssignment_2(t):
    a=1
    b=2
    x = t
    z = b * (1 - t**2/a**2)
    name = 'a parabola (t, b(1 - t^2/a^2))'
    return x, z, name

plotpath2D(MatchingAssignment_2, (-2, 2))
```

```python
def MatchingAssignment_3a(t):
    a=1
    b=2
    x = -a*b/t
    y = t
    name = 'x = -ab/y: (-ab/t, t)'
    return x, y, name

plotpath2D(MatchingAssignment_3a, (1, 2))
```

```python
def MatchingAssignment_3b(t):
    ab=20
    x = -ab/t
    z = t**2
    name = 'complicated way of saying z = -ab/x^3: (-ab/t, t^2)'
    return x, z, name
plotpath2D(MatchingAssignment_3b, (1, 4))
```

```python
def MatchingAssignment_5a(t):
    r = 1
    x = r * np.sinh(t)
    z = r * np.cosh(t)
    name = '(r*sinh(t), t, r*cosh(t))'
    return x, z, name
def MatchingAssignment_5b(t):
    r = 1
    x = r * np.sinh(t)
    z = -r * np.cosh(t)
    name = '(r*sinh(t), t, -r*cosh(t))'
    return x, z, name

multiplot2D((MatchingAssignment_5a, MatchingAssignment_5b), (-2,2))
```
