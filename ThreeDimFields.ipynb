## 3D fields

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

Define a few things to draw 3-D vector fields

```python
def makeGrid(mi=-10,ma=11,step=4):
    return np.meshgrid(np.arange(mi, ma, step),
                      np.arange(mi, ma, step),
                      np.arange(mi, ma, step))
def draw3DVectors(F):
    x, y, z = makeGrid()
    u, v, w = F(x, y, z)
    
    fig = plt.figure(figsize=(10,10))
    ax = fig.gca(projection='3d')
    ax.quiver(x, y, z, u, v, w, length=0.2)
    plt.show()
```

Below, show field $\vec{F} = \left(\begin{array}{c} x \\ y \\ z \end{array}\right)$

```python
def F(x,y,z):
    return x, y, z
draw3DVectors(F)
```

```python
def F2(x,y,z):
    return -y, x, 0
draw3DVectors(F2)
```
